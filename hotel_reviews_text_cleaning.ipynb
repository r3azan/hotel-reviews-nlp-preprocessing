```python
import pandas as pd
import json
```

```python
# Deserializing the Json Text
with open('hotel_reviews.json', 'r') as f:
    data = json.load(f)
```

```python
# pass it to the pandas DataFrame
df = pd.DataFrame(data['reviews'])
df
```

```python
# Lowercasing	all	words
df['review'] = df['review'].str.lower()
df
```

```python
# Removing punctuation marks with regular expresssions
df['review'] = df['review'].str.replace(r'[\!\.\,]','',regex=True)
df
```

```python
# download the stop words data
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_word = set(stopwords.words('english'))
```

```python
def remove_stopwords(text):
  # Tokenizing the text
    words = text.split()      #splits a string into a list, default separator is any whitespace
    filtered_words = [word for word in words if word not in stop_word] #filter list
    return ' '.join(filtered_words)   #join the list of words back into a single string and separate them with a space
```

```python
# Removing	stopwords
df['review'] = df['review'].map(remove_stopwords) #apply remove_stopwords function to each element in the review column
df
```

```python
# Removing emojis with regular expresssions
df['review'] = df['review'].str.replace(r'[^\w\s\']','',regex=True)
df
```
